# KNN

```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
```

```python
x = iris.data #x通常代表題目
y = iris.target #Y通常代表答案
```

```python
X_train , X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1) #test_size整數代表幾題 小數是百分比 random_state= 0~40 不寫每次抓都不一樣 有寫都是固定的
```

```python
knn = KNeighborsClassifier(n_neighbors=3, p=2, weights='distance')  #p代表怎麼計算距離
```

```python
knn.fit(X_train, y_train)
```

```python
knn.predict(X_test)
```

```python
y_test
```

```python
knn.score(X_test,y_test)  #評分
```

Find K

```python
from sklearn import metrics
accuracy = list()
def findK():
    for k in range(1,100):
        knn = KNeighborsClassifier(n_neighbors=k, p=2)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        accuracy.append(metrics.accuracy_score(y_test,y_pred))
    k_range = range(1,100)
    plt.plot(k_range, accuracy)
    plt.show()
```

# 線性回歸

1. 線性回歸: y(x) = w0 + w1x
2. 多項式回歸: y(x) = w0 + w1x^1 + w2x^2 + ... + wnx^n

*   y = a+ b*x +e
*   a: y軸的截距
*   b: 迴歸斜距

# 線性迴歸方程式
*   h(x) = w0 + w1x
*   線性代數: 權重x ax=b
*   將兩邊同乘
*   A^TAx = A^TB
*   x = (A^TA)^-1 * A^TB

```python
def refresh(X,y,w,a):
    dJ = (X.dot(w.T)-y).dot(X)/len(y) #權重公式
    newW = w-a*dJ
    return newW
```

```python
def Polynomial_regression(s,x,y,test_x,test_y):
    # n : n次特徵轉換
    # a : 學習速度參數
    # T : 更新次數
    # w : 權重
    # X : 資料矩陣
    n = s+1
    a = 1
    T = 1000
    X = np.zeros((len(x),n))  #和np.zeros(5) 不一樣
    #產生資料矩陣
    for i in range(n):
        X[:,i] = x**i #次方
    #初始化權重
    w = (np.linalg.inv((X.T).dot(X)).dot(X.T)).dot(y)  #逆矩陣 資料和要是偶數 奇數會發生錯誤
    #紀錄 w產生的預估值
    plot_yy = X.dot(w.T)
    for t in range(T):
        w = refresh(X,y,w,a)
    plot_y = X.dot(w.T)
    #計算相差多少
    error = 0
    for k in range(len(y)):
        error += abs(plot_y[k]-y[k])  #abs絕對值
    error = error / len(y)
    print('訓練錯誤有:',error)
    #產生測試集資料矩陣
    test_X = np.zeros((len(test_x),n))
    #產生資料矩陣
    for i in range(n):
        test_X[:,i] = test_x**i
    pred_error = 0
    pred_y = test_X.dot(w.T)
    #計算測試集的錯誤率
    for k in range(len(pred_y)):
        pred_error += abs(pred_y[k]-test_y[k])
    pred_error = pred_error / len(test_y)
    print('測試錯誤有:',pred_error)
    #繪圖
    plt.rcParams['figure.figsize'] = (10,10)
    plt.plot(x, plot_y,'g-',linewidth=6,label='fitting')
    plt.plot(test_x,pred_y,'g-',linewidth=6)
    plt.plot(x,y,'b-',linewidth=3,label='training data')
    plt.plot(test_x,test_y,'r-',linewidth=3,label='testing data')
    plt.title('TSMC: '+str(n-1)+' feature')
    plt.legend(loc=6, fontsize=20)
    plt.grid()
    plt.show()
    return error,pred_error
```

```python
errorx = np.zeros(20)
errory_train = np.zeros(20)
errory_test = np.zeros(20)
for i in range(1,21):
    errorx[i-1] = i
    errory_train[i-1], errory_test[i-1] = Polynomial_regression(i, x, y, test_x, test_y)
plt.rcParams['figure.figsize'] = (10,10)
print(' ')
plt.title('error')
plt.plot(errorx, errory_train, 'r-o', linewidth=3, markersize=10, label='training error')
plt.plot(errorx, errory_test, 'g-o', linewidth=3, markersize=10, label='testing error')
plt.xlabel('power')
plt.ylabel('error')
plt.legend(loc=6)
plt.grid()
plt.show()
```

# 干擾因子

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)  #隨機種子 隨機抓出來的值一樣
np.random.randn(100) #高斯分布: 變異數為1 平均為0
```

```python
np.random.seed(0)
time = np.arange(1,101)
stock_price = 50+ 2*time + np.random.randn(100)*5 #自定股價
interface = np.random.randn(100)*5  #干擾因子
interface_stock_price = stock_price + interface
plt.figure(figsize=(10,6))
plt.plot(time, stock_price, label='stock price')
plt.plot(time, interface_stock_price, label='interface stock price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend(loc=6)
plt.title('Stock')
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(time.reshape(-1,1), interface_stock_price)
info = model.coef_[0]
info #係數
```

```python
intercept = model.intercept_
intercept #截距
```

```python
print(f"Stock_price = {info:.2f} * time + {intercept:.2f}")   #印出線性回歸方程式
```

```python
from matplotlib.colors import ListedColormap
def plot_decision_regions(X,y,classifier,test_idx=None,resolution=0.02):
    markers = ('s','x','o','^')
    colors = ('red','blue','gray','cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])  #np.unique將重複的值濾掉並進行排序
    x1_min, x1_max = X[:,0].min()-1, X[:,0].max()+1
    x2_min, x2_max = X[:,1].min()-1, X[:,1].max()+1
    xx1, xx2 = np.meshgrid(np.arange(x1_min,x1_max,resolution), np.arange(x2_min,x2_max,resolution))  #meshgrid兩個座標軸的點在平面上劃格線
    z = classifier.predict(np.array([xx1.ravel(),xx2.ravel()]).T)

    z = z.reshape(xx1.shape)

    plt.contourf(xx1,xx2,z,alpha=0.4,cmap=cmap) #畫等高線圖
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y==cl,0], y=X[y==cl,1],alpha=0.6,c=cmap(idx),edgecolors='black',marker=markers[idx],label=cl)
```

```python
plot_decision_regions(X_train_std, y_train['target'].values, classifier=lr)
plt.xlabel('sepal length')
plt.ylabel('pedal length')
plt.legend(loc=6)
plt.show()
```
